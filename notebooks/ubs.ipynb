```
def checa_desempenho(x):
  if x == 'mediano ou  um pouco abaixo da media':
    x = 1
  elif x == 'acima da media' or 'muitoacima da media':
    x = 2
  elif x == 'muito acima da media':
    x = 3
  return x
```

```
notas_sradi = df['sradi'].apply(lambda x : checa_desempenho(x))
notas_srefa = df['srefa'].apply(lambda x : checa_desempenho(x))
notas_sre = df['sre'].apply(lambda x : checa_desempenho(x))
notas_srm = df['srm'].apply(lambda x : checa_desempenho(x))
```
